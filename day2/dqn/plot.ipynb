# Evaluation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = './learning_curves/res.csv'
data = np.genfromtxt(path, delimiter=',')
print(data.shape)
```

```python
steps = data[:, 0] / 1e4
scores = data[:, 1]
```

```python
plt.plot(steps, scores)
plt.title('CartPole-v0: DQN')
plt.xlabel(r'step ($\times 10^4$)')
plt.ylabel('score')
plt.xlim(steps[0], steps[-1])
plt.grid()
```
